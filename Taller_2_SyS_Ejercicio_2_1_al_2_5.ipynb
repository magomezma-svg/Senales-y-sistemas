üîπ**EJERCICIO:**  En el cuarto paso, la evaluaci√≥n de las integrales de convoluci√≥n para los tres casos se deja abierta como un ejercicio. Tenga en cuenta que, en el primer caso, no se superponen, es decir que $y(t)=0$ para $t$ < 0.

$$SOLUCI√ìN$$

Se debe determinar el resultado de $ y(t) = (x*h)(t) $ considerando el desplazamiento de una se√±al sobre la otra y analizar en qu√© intervalos existe superposici√≥n entre ellas.

> En el primer caso no hay superposici√≥n, es decir, $( y(t)=0 $) para $( t<0 $).


- Evaluar paso a paso la integral de convoluci√≥n de dos se√±ales rectangulares causales.  
- Identificar los tres casos:  
  1. Sin traslape (antes del contacto)  
  2. Con traslape (durante la intersecci√≥n)  
  3. Sin traslape (despu√©s de que se separan)  
- Verificar que el resultado anal√≠tico corresponde a la forma triangular esperada.

**Se√±ales consideradas:**  
$[
x(t) = u(t) - u(t - T_1), \qquad
h(t) = u(t) - u(t - T_2)
$]

donde $(u(t)$) es la funci√≥n escal√≥n unitario, y $(T_1, T_2 > 0$) son los anchos de cada rect√°ngulo.

**Definici√≥n de la convoluci√≥n:**
$[
y(t) = \int_{-\infty}^{\infty} x(\tau)\,h(t-\tau)\,d\tau
$]

**Resultado esperado (forma general):**
$[
y(t) =
\begin{cases}
0, & t < 0, \\
t, & 0 \le t \le T_{\min}, \\
T_{\min}, & T_{\min} \le t \le T_{\max}, \\
T_1 + T_2 - t, & T_{\max} \le t \le T_1 + T_2, \\
0, & t \ge T_1 + T_2,
\end{cases}
$]
donde $(T_{\min} = \min(T_1, T_2)$) y $(T_{\max} = \max(T_1, T_2)$).

- En el **primer caso** (sin traslape, $(t<0$)) no hay √°rea com√∫n entre las se√±ales ‚Üí $(y(t)=0$).  
- En el **segundo caso**, la integral crece o se mantiene constante seg√∫n la superposici√≥n.  
- En el **tercer caso**, al separarse las se√±ales, la integral decrece hasta volver a cero.  
El resultado es una **forma triangular** que representa el √°rea de intersecci√≥n variable entre las se√±ales.

### 1) Planteamiento
Trabajo con dos rect√°ngulos **causales** unitarios (altura 1), de anchos $(T_1$) y $(T_2$):
$[
x(t)=u(t)-u(t-T_1),\qquad h(t)=u(t)-u(t-T_2).
$]
La convoluci√≥n es:
$[
y(t)=(x*h)(t)=\int_{-\infty}^{\infty} x(\tau)\,h(t-\tau)\,d\tau.
$]
Como ambas se√±ales valen 1 s√≥lo en sus ventanas, la integral **se reduce a medir el largo del traslape** entre los intervalos:
$[
\tau\in[0,\,T_1]\quad \cap \quad \tau\in[t-T_2,\,t].
$]

---

### 2) Soporte de cada t√©rmino
- $(x(\tau)=1$) si $(\tau\in[0,\,T_1]$), y 0 fuera.
- $(h(t-\tau)=1$) si $(t-\tau\in[0,\,T_2]\ \Leftrightarrow\ \tau\in[t-T_2,\,t]$); 0 fuera.

Entonces:
$[
y(t)=\int_{\max(0,\,t-T_2)}^{\min(T_1,\,t)} 1\; d\tau
= \big[\min(T_1,t)-\max(0,t-T_2)\big]_+,
$]
donde $([z]_+=\max(z,0)$) asegura que, si no hay traslape, el resultado sea 0.

---

### 3) An√°lisis por tramos (los ‚Äútres casos‚Äù del cuaderno)

Defino $(T_{\min}=\min(T_1,T_2)$) y $(T_{\max}=\max(T_1,T_2)$).

**Caso 1 ‚Äî Sin traslape inicial (no superposici√≥n):**  
Para $(t<0$) todav√≠a no se ‚Äútocan‚Äù los intervalos, as√≠ que:
$[
y(t)=0 \quad \text{(este es el primer caso que menciona el cuaderno).}
$]

**Caso 2 ‚Äî Con traslape (intersecci√≥n variable):**  
A medida que $(t$) avanza, la intersecci√≥n crece, puede mantenerse constante (si $(T_1\neq T_2$)) y luego decrece:

- **Creciente:** $(0\le t\le T_{\min}$)  
  $[
  y(t)=t.
  $]
  (El solapamiento aumenta linealmente desde 0 hasta \(T_{\min}\).)

- **Meseta (solo si $(T_1\neq T_2$)):** $(T_{\min}\le t\le T_{\max}$)  
  $[
  y(t)=T_{\min}.
  $]
  (El solapamiento queda limitado por el rect√°ngulo m√°s corto.)

- **Decreciente:** $(T_{\max}\le t\le T_1+T_2$)  
  $[
  y(t)=T_1+T_2 - t.
  $]
  (El solapamiento disminuye linealmente hasta desaparecer.)

**Caso 3 ‚Äî Sin traslape final:**  
Para $(t\ge T_1+T_2$) los intervalos ya se separaron por completo:
$[
y(t)=0.
$]

---

### 4) Resultado pieza a pieza (forma cerrada)
$[
y(t)=
\begin{cases}
0, & t<0,\\[4pt]
t, & 0\le t\le T_{\min},\\[4pt]
T_{\min}, & T_{\min}\le t\le T_{\max},\\[4pt]
T_1+T_2 - t, & T_{\max}\le t\le T_1+T_2,\\[4pt]
0, & t\ge T_1+T_2.
\end{cases}
$]

**Observaci√≥n:** si $(T_1=T_2=T$), la ‚Äúmeseta‚Äù desaparece y queda un **tri√°ngulo is√≥sceles**:
$[
y(t)=
\begin{cases}
0, & t<0,\\
t, & 0\le t\le T,\\
2T-t, & T\le t\le 2T,\\
0, & t\ge 2T.
\end{cases}
$]

---

### 5) Escalado por alturas (caso general)
Si las alturas no son 1 (por ejemplo $(x$) con altura $(A_1$) y $(h$) con altura $(A_2$)), el resultado se **multiplica por $(A_1A_2$)**:
$[
y_{\text{general}}(t) = A_1 A_2 \cdot y_{\text{unitario}}(t).
$]

---

### 6) Verificaci√≥n num√©rica (opcional)
Con muestreo fino $(\Delta t$), la convoluci√≥n discreta cumple:
$[
y_{\text{num}}[n]\ \approx\ \sum_k x[k]\;h[n-k]\;\Delta t,
$]
y debe **coincidir** con la curva anal√≠tica por tramos.  
(En el c√≥digo de abajo hago esta verificaci√≥n y grafico ambas curvas).



```python
import numpy as np
import matplotlib.pyplot as plt

# ----- par√°metros (puedes cambiarlos) -----
T1 = 2.0   # ancho del primer rect√°ngulo
T2 = 3.0   # ancho del segundo rect√°ngulo
A1 = 1.0   # altura x(t)
A2 = 1.0   # altura h(t)  (si A1,A2 != 1, y(t) se escala por A1*A2)

# ----- eje temporal continuo (muestreo fino para visualizar) -----
dt = 1e-3
t = np.arange(-1.0, T1+T2+1.0, dt)  # tiempo para graficar

# ----- definiciones rectangulares causales -----
def rect_causal(t, T, A=1.0):
    """A*(u(t)-u(t-T)) -> 1 entre [0, T], 0 fuera."""
    return A * ((t >= 0) & (t <= T)).astype(float)

x = rect_causal(t, T1, A1)
h = rect_causal(t, T2, A2)

# ----- y(t) por f√≥rmula de traslape (anal√≠tica pieza a pieza) -----
Tmin, Tmax = min(T1, T2), max(T1, T2)

y = np.zeros_like(t)
# Caso crec: 0 <= t <= Tmin -> y = t
mask1 = (t >= 0) & (t <= Tmin)
y[mask1] = t[mask1]

# Caso meseta (si T1 != T2): Tmin <= t <= Tmax -> y = Tmin
mask2 = (t >= Tmin) & (t <= Tmax)
y[mask2] = Tmin

# Caso decreciente: Tmax <= t <= T1+T2 -> y = T1 + T2 - t
mask3 = (t >= Tmax) & (t <= T1+T2)
y[mask3] = (T1 + T2) - t[mask3]

# Escala por alturas si A1,A2 != 1
y *= (A1 * A2)

# ----- verificaci√≥n num√©rica por ‚Äúconvoluci√≥n‚Äù discreta (CORREGIDA) -----
x_d = rect_causal(t, T1, A1)
h_d = rect_causal(t, T2, A2)

y_num = np.convolve(x_d, h_d) * dt           # *dt aproxima la integral
t_start = t[0] + t[0]                         # inicio del soporte (suma de los inicios)
t_num = t_start + np.arange(len(y_num)) * dt  # eje que coincide con y_num

# ----- gr√°ficas -----
plt.figure(figsize=(11,7))
plt.subplot(3,1,1)
plt.plot(t, x, label='x(t) = rect(0..T1)')
plt.grid(True); plt.legend(); plt.ylabel('amplitud')

plt.subplot(3,1,2)
plt.plot(t, h, 'C1', label='h(t) = rect(0..T2)')
plt.grid(True); plt.legend(); plt.ylabel('amplitud')

plt.subplot(3,1,3)
plt.plot(t, y, 'C2', lw=2, label='y(t) (anal√≠tica por traslape)')
plt.plot(t_num, y_num, 'k--', alpha=0.6, label='y_num ‚âà x*h (discreta)')
plt.grid(True); plt.legend(); plt.xlabel('t [s]'); plt.ylabel('y(t)')
plt.suptitle('Convoluci√≥n de dos rect√°ngulos causales ‚Äî tres casos')
plt.tight_layout()
plt.show()
```

# 2.2 Consultar y realizar los ejercicios propuestos en el cuaderno de Colab Respuesta impulso.

  **A. Compare la se√±al de salida obtenida al resolver la EDO con la obtenida mediante la convoluci√≥n. ¬øSon iguales?**

  $$SOLUCI√ìN$$

  üîπEl sistema analizado en el cuaderno es un sistema **lineal e invariante en el tiempo (LTI)** descrito por la ecuaci√≥n diferencial:

$[
\frac{dy(t)}{dt} + a\,y(t) = x(t),
$]

con condici√≥n inicial nula \(y(0^-)=0\).  
La funci√≥n de transferencia en el dominio de Laplace es:

$[
H(s) = \frac{1}{s + a}.
$]

La **respuesta al impulso** del sistema es entonces:

$[
h(t) = e^{-a t}\,u(t),
$]

y la **salida por convoluci√≥n** se obtiene como:

$[
y(t) = (h * x)(t) = \int_0^t e^{-a(t - \tau)}\,x(\tau)\,d\tau.
$]

---

### **üîπ Procedimiento**
1. Se simula el sistema usando la ecuaci√≥n diferencial $(y'(t) + a y(t) = x(t)$) con una entrada $(x(t)$) (por ejemplo, un escal√≥n o un pulso rectangular).  
2. Se calcula la salida equivalente aplicando la **convoluci√≥n** de $(x(t)$) con $(h(t) = e^{-a t}u(t)$).  
3. Finalmente, se comparan ambas se√±ales $(y(t)$) obtenidas por los dos m√©todos.

---

### **üîπ Observaciones**
- Para condiciones iniciales nulas ($(y(0^-)=0$)), las dos respuestas son **id√©nticas**, ya que la convoluci√≥n reproduce exactamente la soluci√≥n particular de la EDO.  
- La peque√±a diferencia num√©rica entre ambas curvas solo se debe al paso de muestreo ($(\Delta t$)) usado en la integraci√≥n.  
- La forma de $(y(t)$) depende de la entrada:
  - Si $(x(t) = u(t)$):  
    $[
    y(t) = \frac{1 - e^{-a t}}{a}.
    $]
  - Si $(x(t)$) es un pulso de duraci√≥n $(T_p$): la salida crece exponencialmente mientras el pulso est√° activo y decae cuando termina.

---
La salida obtenida por **resoluci√≥n directa de la EDO** y la obtenida mediante la **convoluci√≥n con la respuesta al impulso** son **iguales** siempre que:
- El sistema sea **LTI** (lineal e invariante en el tiempo), y  
- Las **condiciones iniciales sean nulas**.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lsim, TransferFunction

# -------- Par√°metros del sistema (EDO: y' + a y = x) --------
a = 2.0                 # puedes cambiar a
dt = 1e-3
T  = 3.0
t  = np.arange(0, T+dt, dt)

# -------- Entradas de prueba --------
# 1) Escal√≥n u(t)
u_step = np.ones_like(t)

# 2) Pulso rectangular de duraci√≥n Tp (desde t=0 hasta t=Tp)
Tp = 0.6
u_pulse = (t >= 0) & (t <= Tp)
u_pulse = u_pulse.astype(float)

# -------- 1) Salida resolviendo la EDO (simulaci√≥n de H(s)=1/(s+a)) --------
H = TransferFunction([1.0], [1.0, a])   # numerador, denominador
tout1, y_step_edo, _  = lsim(H, U=u_step,  T=t)     # a escal√≥n
tout2, y_pulse_edo, _ = lsim(H, U=u_pulse, T=t)     # a pulso

# -------- 2) Salida por convoluci√≥n con h(t)=e^{-a t} u(t) --------
h = np.exp(-a*t)                         # respuesta al impulso (u(t) impl√≠cito porque t>=0)
# Convoluci√≥n continua aproximada: y ‚âà (x * h) * dt
y_step_conv  = np.convolve(u_step,  h)[:len(t)]  * dt
y_pulse_conv = np.convolve(u_pulse, h)[:len(t)] * dt

# -------- Comparaci√≥n num√©rica --------
err_step  = np.max(np.abs(y_step_edo  - y_step_conv))
err_pulse = np.max(np.abs(y_pulse_edo - y_pulse_conv))
print(f"Error m√°x (escal√≥n): {err_step:.3e}")
print(f"Error m√°x (pulso)  : {err_pulse:.3e}")

# -------- Gr√°ficas --------
plt.figure(figsize=(11,6))
plt.subplot(2,2,1)
plt.plot(t, u_step); plt.grid(True); plt.title("Entrada 1: u(t)")
plt.subplot(2,2,2)
plt.plot(t, u_pulse); plt.grid(True); plt.title(f"Entrada 2: pulso (Tp={Tp}s)")

plt.subplot(2,2,3)
plt.plot(t, y_step_edo, label="EDO");
plt.plot(t, y_step_conv, '--', label="Convoluci√≥n")
plt.grid(True); plt.title("Salida a escal√≥n"); plt.xlabel("t [s]"); plt.legend()

plt.subplot(2,2,4)
plt.plot(t, y_pulse_edo, label="EDO");
plt.plot(t, y_pulse_conv, '--', label="Convoluci√≥n")
plt.grid(True); plt.title("Salida a pulso"); plt.xlabel("t [s]"); plt.legend()

plt.tight_layout(); plt.show()
```

**B. Compruebe la soluci√≥n $h(t)$  de la EDO cuando $x(t)$ de manera manual. Tener en cuenta que $\frac{d}{dt} \epsilon(t) = \delta(t)$.**

$$SOLUCI√ìN$$

Comprobar manualmente la soluci√≥n $( h(t) $) de la EDO cuando la entrada es $( x(t)=\delta(t) $). **
Recordar que $( \frac{d}{dt}\,\varepsilon(t) = \delta(t) $), donde $( \varepsilon(t) $) es la funci√≥n escal√≥n (Heaviside).

---

### **üîπ Ecuaci√≥n diferencial base**
El sistema se describe por:
$[
\frac{dy(t)}{dt} + a\,y(t) = x(t),
$]
y para obtener la **respuesta al impulso**, hacemos:
$[
x(t) = \delta(t), \qquad y(t) = h(t).
$]

Sustituyendo:
$[
\frac{dh(t)}{dt} + a\,h(t) = \delta(t),
$]
con condici√≥n inicial nula \( h(0^-) = 0 \).

---

### **üîπ Resoluci√≥n paso a paso**

#### a) Soluci√≥n para \( t > 0 \)
Cuando $( t > 0 $), la delta desaparece ($(\delta(t)=0$)), as√≠ que la EDO homog√©nea es:
$[
\frac{dh(t)}{dt} + a\,h(t) = 0.
$]
Su soluci√≥n general es:
$[
h(t) = C\,e^{-a t}, \quad t>0.
$]

#### b) Determinar la constante $(C$)
Integramos la EDO original alrededor de $( t=0 $):
$[
\int_{0^-}^{0^+} \left(\frac{dh(t)}{dt} + a\,h(t)\right) dt = \int_{0^-}^{0^+} \delta(t)\,dt.
$]

Evaluando los t√©rminos:
$[
[h(0^+) - h(0^-)] + a\int_{0^-}^{0^+}h(t)dt = 1.
$]

Como $(h(0^-)=0$) y el segundo t√©rmino es despreciable (integral infinitesimal), queda:
$[
h(0^+) = 1.
$]

Entonces:
$[
C = 1.
$]

---

### **üîπ Resultado final**
$[
\boxed{h(t) = e^{-a t}\,u(t)}
$]

Es decir, la **respuesta al impulso** es un exponencial decreciente que comienza en 1 en $(t=0^+$).

---

### **üîπ Interpretaci√≥n**
- $(h(t)$) describe c√≥mo responde el sistema a un impulso unitario.  
- El par√°metro $(a$) controla la **velocidad de decaimiento**:  
  - Si $(a>0$), el sistema es **estable** (la respuesta se aten√∫a).  
  - Si $(a=0$), el sistema es **integrador** (salida constante).  
  - Si $(a<0$), el sistema es **inestable** (crece exponencialmente).  
- Adem√°s, como $( \frac{d}{dt}\varepsilon(t) = $delta(t) $), tambi√©n se puede comprobar que:
  $[
  h(t) = $(\frac{d}{dt}\big(e^{-a t}\varepsilon(t)\big )$,
  $]
  lo que confirma que la derivada del escal√≥n exponencial produce la misma $(h(t)$).

---

La soluci√≥n manual demuestra que la **respuesta al impulso del sistema** definido por  
$(\frac{dy(t)}{dt} + a\,y(t) = x(t)$)  
es $(h(t)=e^{-a t}u(t)$), coincidiendo con la obtenida mediante la transformada de Laplace y con la que usamos para la convoluci√≥n en el subpunto anterior.


**C. Comprobar la soluci√≥n de la integral de convoluci√≥n de manera manual. Tener en cuenta las funciones Heaviside.**

$$SOLUCI√ìN$$

Parto del sistema LTI de primer orden del cuaderno:
$[
\frac{dy(t)}{dt} + a\,y(t) = x(t),\qquad a>0,\ \ y(0^-)=0,
$]
cuya **respuesta al impulso** es
$[
h(t)=e^{-a t}\,u(t).
$]
Por definici√≥n,
$[
y(t)=(h*x)(t)=\int_{-\infty}^{\infty} h(t-\tau)\,x(\tau)\,d\tau
          =\int_{0}^{t} e^{-a\,(t-\tau)}\,x(\tau)\,d\tau,
$]
donde el l√≠mite inferior se vuelve 0 porque \(h(t)=0\) para \(t<0\).

---

### 1) Entrada escal√≥n $(x(t)=u(t)$)
$[
y(t)=\int_{0}^{t} e^{-a\,(t-\tau)}\cdot 1 \; d\tau
     = \left[-\frac{1}{a} e^{-a\,(t-\tau)}\right]_{\tau=0}^{\tau=t}
     = \frac{1-e^{-a t}}{a}.
$]

**Con Heaviside (para dejarlo pieza a pieza):**
$[
\boxed{\,y(t)=\frac{1-e^{-a t}}{a}\,u(t)\,}
$]
(Es cero para \(t<0\) y vale \((1-e^{-a t})/a\) para \(t\ge 0\).)

---

### 2) Entrada pulso rectangular $(x(t)=u(t)-u(t-T_p)$)
Es un ‚Äúencendido‚Äù entre $(0$) y $(T_p$). Uso linealidad y corrimiento con Heaviside:

$[
\begin{aligned}
y(t)
&=(h*u)(t)\;-\;\big(h*u(t-T_p)\big)(t) \\
&=\frac{1-e^{-a t}}{a}\,u(t)\;-\;\frac{1-e^{-a\,(t-T_p)}}{a}\,u(t-T_p).
\end{aligned}
$]

**Forma pieza a pieza (m√°s clara):**
$[
y(t)=
\begin{cases}
0, & t<0,\\[4pt]
\dfrac{1-e^{-a t}}{a}, & 0\le t < T_p,\\[8pt]
\dfrac{e^{-a\,(t-T_p)} - e^{-a t}}{a}, & t\ge T_p.
\end{cases}
$]

Interpretaci√≥n r√°pida:
- Mientras el pulso est√° ‚Äúencendido‚Äù ($(0\le t<T_p$)), la salida **sube** como el escal√≥n filtrado.
- Cuando el pulso se apaga ($(t\ge T_p$)), la salida **decay√≥** como diferencia de dos exponenciales.

---

### 3) Entrada cualquiera escrita con Heaviside
Si puedo escribir $(x(t)$) como suma de trozos con Heaviside (p. ej. sumas de $(u(t-t_i)$) y pulsos $(u(t-t_i)-u(t-t_{i+1})$)), entonces:
- Convolucionar con $(h$) es aplicar las f√≥rmulas anteriores a cada trozo,
- y **sumar** los resultados, cuidando los corrimientos $(u(t-t_i)$).

---


- La soluci√≥n manual de la integral de convoluci√≥n para el sistema $(h(t)=e^{-a t}u(t)$) **coincide** con la salida que obtuvimos resolviendo la EDO.  
- Las **funciones Heaviside** permiten escribir la respuesta **por tramos** de forma compacta:
  $[
  (h*u)(t)=\frac{1-e^{-a t}}{a}\,u(t)\quad \text{y}\quad
  (h*\big[u(t)-u(t-T_p)\big])(t)=\frac{1-e^{-a t}}{a}\,u(t)-\frac{1-e^{-a (t-T_p)}}{a}\,u(t-T_p).
  $]
- Esta verificaci√≥n cierra el ciclo: **EDO ‚Üî h(t) ‚Üî convoluci√≥n**, todo consistente.


# 2.3 Demuestre si los siguientes sistemas de la forma $y = H{x}$, son sistemas lineales e invariantes en el tiempo (SLIT) (simule los sistemas en Python):
a) $y[n] = \dfrac{x[n]}{3} + 2x[n - 1] - y[n - 1]$  

b) $y[n] = \displaystyle\sum_{k=-\infty}^{n} x^2[k]$  

c) $y[n] = \text{median}(x[n])$, donde *median* es la funci√≥n mediana sobre una ventana de tama√±o $3$.  

d) $y(t) = A\,x(t) + B, \quad A,B \in \mathbb{R}$


$$SOLUCI√ìN$$

Un sistema es **lineal** si cumple las dos propiedades:

1. **Homogeneidad:**  
   Si la entrada se multiplica por una constante, la salida tambi√©n se multiplica por la misma constante.  
   $[
   \mathcal{H}\{\alpha x[n]\} = \alpha\,\mathcal{H}\{x[n]\}
   $]

2. **Aditividad:**  
   La respuesta a la suma de entradas es la suma de las respuestas individuales.  
   $[
   \mathcal{H}\{x_1[n] + x_2[n]\} = \mathcal{H}\{x_1[n]\} + \mathcal{H}\{x_2[n]\}
   $]

Y es **invariante en el tiempo (TI)** si al desplazar la entrada en el tiempo, la salida tambi√©n se desplaza igual:
$[
\mathcal{H}\{x[n-n_0]\} = y[n-n_0]
$]

---

## **Sistema 1:**
$[
y[n] = \frac{1}{3}x[n] + 2x[n-1] - y[n-1]
$]

### **An√°lisis paso a paso**

**a) Linealidad:**

- No hay potencias, productos ni t√©rminos no lineales.  
- Si $(x_1[n] \to y_1[n]$) y $(x_2[n] \to y_2[n]$):  
  $[
  \mathcal{H}\{\alpha x_1 + \beta x_2\} = \frac{1}{3}(\alpha x_1+\beta x_2) + 2(\alpha x_1[n-1]+\beta x_2[n-1]) - (\alpha y_1[n-1]+\beta y_2[n-1])
  $]
  que es igual a $(\alpha y_1[n] + \beta y_2[n]$).  
  Por tanto, **es lineal**.

**b) Invariancia en el tiempo:**

Si desplazo la entrada \(x[n-n_0]\), la ecuaci√≥n se convierte en:
\[
y[n-n_0] = \frac{1}{3}x[n-n_0] + 2x[n-1-n_0] - y[n-1-n_0],
\]
que tiene la misma forma.  
Por tanto, **s√≠ es invariante en el tiempo.**


El sistema es **lineal e invariante en el tiempo (SLIT)**.  
Corresponde a un **sistema IIR** lineal con coeficientes constantes.

---

##  **Sistema 2:**
$[
y[n] = \sum_{k=-\infty}^{n} x^2[k]
$]

### **a) Linealidad**

- Contiene un t√©rmino $(x^2[k]$), que rompe la propiedad de homogeneidad.  
- Si \(x[n]\) se multiplica por $(\alpha$):
  $[
  y_{\text{nuevo}}[n] = \sum (\alpha x[k])^2 = \alpha^2 \sum x^2[k] \neq \alpha y[n].
  $]
  ‚áí No cumple linealidad.

### **b) Invariancia en el tiempo**

Desplazo la entrada \(x[n-n_0]\):
$[
\mathcal{H}\{x[n-n_0]\} = \sum_{k=-\infty}^{n} x^2[k-n_0].
$]
Cambiando variable \(m = k - n_0\):
$[
\sum_{m=-\infty}^{n-n_0} x^2[m] = y[n-n_0].
$]
‚áí S√≠ cumple **invariancia en el tiempo**.


El sistema **no es lineal**, pero **s√≠ es invariante en el tiempo**.  
Por tanto, **no es SLIT**.

---

## **Sistema 3:**
$[
y[n] = \operatorname{median}\{x[n-1], x[n], x[n+1]\}
$]

### **a) Linealidad**

- La funci√≥n **mediana** no cumple ni homogeneidad ni aditividad.

**Ejemplo r√°pido:**
- $(x_1[n] = [0, 100, 0]$) ‚Üí mediana = 0  
- $(x_2[n] = [0, 0, 100]$) ‚Üí mediana = 0  
- $(x_1[n] + x_2[n] = [0, 100, 100]$) ‚Üí mediana = 100  
Entonces:
$[
\mathcal{H}\{x_1+x_2\} \neq \mathcal{H}\{x_1\} + \mathcal{H}\{x_2\}.
$]
‚áí No es lineal.

### **b) Invariancia en el tiempo**

El operador mediana se aplica siempre sobre la misma ventana relativa a $(n$):  
si la entrada se desplaza, la ventana tambi√©n se desplaza igual.  
Por tanto, **s√≠ es invariante en el tiempo.**


**No lineal, pero s√≠ TI.**  
**No es SLIT.**

---

## **4Ô∏èSistema 4:**
$[
y(t) = A\,x(t) + B,\quad A,B\in\mathbb{R}
$]

### **a) Linealidad**

- Si $(B=0$):  
  $[
  y(t) = A x(t)
  $]
  Cumple homogeneidad y aditividad ‚áí **lineal**.

- Si $(B\neq0$):  
  $(\mathcal{H}\{0\}=B\neq0$) ‚áí no cumple homogeneidad.  
  Por tanto, **no lineal**.

### **b) Invariancia en el tiempo**

Desplazo la entrada \(x(t-t_0)\):
$[
\mathcal{H}\{x(t-t_0)\} = A\,x(t-t_0) + B = y(t-t_0).
$]
‚áí Siempre **invariante en el tiempo**.


- Si \(B=0\) ‚Üí **S√≠ es SLIT**.  
- Si \(B\neq0\) ‚Üí **No es lineal**, aunque sigue siendo **TI**.

$$SOLUCI√ìN$$

Hallar la salida $( y[n] $) de un sistema lineal e invariante en el tiempo (SLIT) ante la entrada:

$[

$]

con respuesta al impulso:

h[n] = \{1,\, -2,\, 0^{\dagger},\, 1,\, -2\},

### **1Parte 1 ‚Äî Convoluci√≥n con la respuesta al impulso ($h[n]$)**

El sistema es **lineal e invariante en el tiempo**, por lo tanto la salida se obtiene aplicando la **convoluci√≥n discreta**:

$[
y[n] = (x * h)[n] = \sum_{k=-\infty}^{\infty} x[k]\,h[n-k].
$]

Cada valor de $(y[n]$) se calcula multiplicando punto a punto $(x[k]$) con la versi√≥n invertida y desplazada de $(h[k]$).

El origen de $(x[n]$) est√° en la posici√≥n 2 (el valor $(-3$))

$[
y[n] = \{-15,\; 35,\; -13,\; -9,\; \mathbf{40},\; -16,\; -9,\; 7,\; -3,\; -15,\; 2\}.
$]

El valor en **negrita** corresponde a $(n=0$).

Para esta parte, primero se obtiene la **respuesta al impulso** a partir del escal√≥n:

$ [

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Se√±ales base (definidas a partir del enunciado del problema)
# x[n] = {-15, 5, -3‚Ä†, 0, 5, 7, -1} donde ‚Ä† es n=0
x_vals = [-15, 5, -3, 0, 5, 7, -1]
x0 = 2  # √çndice donde se encuentra n=0 en x_vals

# h[n] = {1, -2, 0‚Ä†, 1, -2} donde ‚Ä† es n=0
h_vals = [1, -2, 0, 1, -2]
h0 = 2  # √çndice donde se encuentra n=0 en h_vals

# Salida y[n] = x[n] * h[n] (calculada en la soluci√≥n del texto)
y1_vals = [-15, 35, -13, -9, 40, -16, -9, 7, -3, -15, 2]
n1_offset = x0 + h0 # Offset del origen para y1
n1 = np.arange(-n1_offset, len(y1_vals) - n1_offset)
y1 = np.array(y1_vals)

# s[n] = {-1, 6, -10, 3‚Ä†, 1, -10, 2, 5} donde ‚Ä† es n=0
s_vals = [-1, 6, -10, 3, 1, -10, 2, 5]
s0 = 3 # √çndice donde se encuentra n=0 en s_vals

# h[n] derivada de s[n] = {-1, 7, -16, 13‚Ä†, -2, -11, 12, 3} donde ‚Ä† es n=0
h_derived_vals = [-1, 7, -16, 13, -2, -11, 12, 3]
h_derived_0 = 3 # √çndice donde se encuentra n=0 en h_derived_vals

# Salida y[n] = x[n] * h_derived[n] (calculada en la soluci√≥n del texto)
y2_vals = [15, -110, 278, -296, 138, 144, -259, -6, 76, -91, -15, 110, 9, -3]
n2_offset = x0 + h_derived_0 # Offset del origen para y2
n2 = np.arange(-n2_offset, len(y2_vals) - n2_offset)
y2 = np.array(y2_vals)

# ===============================
# GRAFICACI√ìN DE LAS SE√ëALES
# ===============================

# ---- Se√±ales base para plotear
x = np.array(x_vals)
h = np.array(h_vals)

# Graficar x[n], h[n], y[n]
fig, axs = plt.subplots(3, 1, figsize=(9, 8))

# --- x[n]
axs[0].stem(np.arange(-x0, len(x)-x0), x, basefmt=" ")
axs[0].set_title("Entrada x[n]")
axs[0].set_xlabel("n")
axs[0].set_ylabel("Amplitud")
axs[0].grid(True)

# --- h[n]
axs[1].stem(np.arange(-h0, len(h)-h0), h, basefmt=" ")
axs[1].set_title("Respuesta al impulso h[n]")
axs[1].set_xlabel("n")
axs[1].set_ylabel("Amplitud")
axs[1].grid(True)

# --- y[n]
axs[2].stem(n1, y1, basefmt=" ") # Usamos y1 aqu√≠
axs[2].set_title("Salida y[n] = x[n] * h[n]")
axs[2].set_xlabel("n")
axs[2].set_ylabel("Amplitud")
axs[2].grid(True)

plt.tight_layout()
plt.show()

# ===============================
# Comparaci√≥n: y[n] con h dada vs con h derivada de s
# ===============================
plt.figure(figsize=(9,4))
plt.stem(n1, y1, linefmt='C0-', markerfmt='o', basefmt=' ', label="y[n] con h dada")
plt.stem(n2, y2, linefmt='C1--', markerfmt='s', basefmt=' ', label="y[n] con h derivada de s")
plt.title("Comparaci√≥n de salidas: h[n] dada vs derivada del escal√≥n s[n]")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 2.5 Sea la se√±al Gaussiana $x(t) = e^{-a t^2}$ con a ‚àà R+, el sistema A con relac√≥n entrada-salida $y_A(t) = x^2(t)$, y el sistema lineal e invariante con el tiempo B con respuesta al impulso $h_B(t) = B\,e^{-b t^2}$:
**a)** Encuentre la salida del sistema en serie $x(t) ‚Üí hB(t) ‚Üí yA(t) ‚Üí y(t)$

**b)** Encuentre la salida del sistema en serie $x(t) ‚Üí yA(t) ‚Üí hB(t) ‚Üí y(t)$.

$$SOLUCI√ìN$$

Sea la se√±al:
$[
x(t)=e^{-a t^2}, \quad a>0
$]
El sistema **A** tiene la relaci√≥n:
$[
y_A(t)=x^2(t)=e^{-2a t^2}.
$]
El sistema **B** es LTI con respuesta al impulso:
$[
h_B(t)=B\,e^{-b t^2}.
$]

Se pide:
1. **a)** Salida del sistema en serie $(x(t)\rightarrow h_B(t)\rightarrow y_A(t)\rightarrow y(t)$).  
2. **b)** Salida del sistema en serie $(x(t)\rightarrow y_A(t)\rightarrow h_B(t)\rightarrow y(t)$).

---



#### **Caso a)**
Primero se aplica el sistema $(B$) y luego el sistema $(A$).

$[
y(t)=y_A(t)=\big[(x*h_B)(t)\big]^2
$]
donde la convoluci√≥n entre dos Gaussianas $(e^{-a t^2}$) y $(e^{-b t^2}$) produce otra Gaussiana:
$[
(x*h_B)(t)=C\,e^{-\frac{a b}{a+b}t^2},\quad C=\sqrt{\frac{\pi}{a+b}}\,B.
$]
Por tanto:
$[
y(t)=C^2\,e^{-\frac{2 a b}{a+b}t^2}.
$]

#### **Caso b)**
Primero se aplica el sistema \(A\) y luego \(B\):
$[
y(t)= (y_A*h_B)(t) = \int e^{-2a\tau^2}\,B e^{-b(t-\tau)^2}\,d\tau.
$]
El resultado es nuevamente Gaussiano:
$[
y(t)=B\sqrt{\frac{\pi}{2a+b}}\,e^{-\frac{2ab}{2a+b}t^2}.
$]

---

En ambos casos la salida es una funci√≥n Gaussiana $(e^{-k t^2}$),  
pero el **par√°metro de dispersi√≥n** cambia seg√∫n el orden de los sistemas:
- En el **caso a)** la atenuaci√≥n depende de $(\frac{2ab}{a+b}$).
- En el **caso b)** depende de $(\frac{2ab}{2a+b}$).

Para comprobarlo num√©ricamente se simulan las convoluciones en Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Par√°metros
a = 1.0
b = 0.8
B = 1.0
t = np.linspace(-5, 5, 1000)

# Se√±al de entrada y sistemas
x = np.exp(-a * t**2)
hB = B * np.exp(-b * t**2)

# --- Caso (a): x ‚Üí hB ‚Üí yA
conv_a = convolve(x, hB, mode='same') * (t[1]-t[0])
y_a = conv_a**2

# --- Caso (b): x ‚Üí yA ‚Üí hB
yA = x**2
conv_b = convolve(yA, hB, mode='same') * (t[1]-t[0])
y_b = conv_b

# --- Gr√°ficas
plt.figure(figsize=(10,7))
plt.subplot(3,1,1)
plt.plot(t, x, label='x(t)=e^{-a t¬≤}')
plt.title('Entrada Gaussiana y sistemas A y B')
plt.grid(True); plt.legend()

plt.subplot(3,1,2)
plt.plot(t, hB, label='h_B(t)=B e^{-b t¬≤}', color='orange')
plt.grid(True); plt.legend()

plt.subplot(3,1,3)
plt.plot(t, y_a, label='Salida caso (a)', color='C2')
plt.plot(t, y_b, '--', label='Salida caso (b)', color='C1')
plt.grid(True); plt.legend()
plt.xlabel('t'); plt.tight_layout()
plt.show()
```
